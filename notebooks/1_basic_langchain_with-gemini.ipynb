#### langchain Model - Gemini
#### DOTENV - Load environment variables from a .env file

### Load environment variables from .env file

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
response = llm.invoke("What is the capital of India?")
print(response.content)
```

### Connect with OpenAI model

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
res = llm.invoke("Explain the concept of LangChain in simple terms.")
print(res.content)
```

#### Static Prompt

```python
prompts = [
    ("system", "you are a AI Engineer."),    #### 1st argument is type : "system : means system message", "user: means human message", "ai: means assistant message"
    ("user", "how to use langchain?")
]

res = llm.invoke(prompts)
print(res.content)
```
