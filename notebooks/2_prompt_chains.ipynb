```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts = [
    {"role": "system", "content": "You are a Javascript developer."},
    {"role": "user", "content": "Difference between filter and map?"},
]
```

```python
res = llm.invoke(prompts)
print(res.content)
```

## Prompt Templates - Dynamic Prompts

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompts = ChatPromptTemplate.from_messages(
    [
    {"role": "system", "content": "You are a translator and you translate input into {language}."},
    {"role": "user", "content": "{query}"},
]
)
```

```python
final_prompts = prompts.format_prompt(language="Hindi", query="Hello, how are you?")
res = llm.invoke(final_prompts)
print(res.content)
```

### Chains - Runnable

```python
final_prompts1 = prompts.invoke({"language": "French", "query": "Hello, how are you?"})
```

```python
res = llm.invoke(final_prompts1.messages)
print(res.content)
```

```python
# Or we can directly do this
# res = llm.invoke(prompts.invoke({"language": "Hinglish", "query": "Hello, how are you?"}).messages)
res = llm.invoke(prompts.invoke({"language": "Hinglish", "query": "Hello, how are you?"}))
print(res.content)
```

### Chains


```python
chains = prompts | llm
chains
```

```python
res = chains.invoke({"language": "Javascript", "query": "Hello, how are you?"})
print(res.content)
```

```python
### to parse output so that we can use it in next step of chain, we can use output parser
from langchain_core.output_parsers import StrOutputParser
out = StrOutputParser()
```

```python
chains = prompts | llm | out
res = chains.invoke({"language": "Russian", "query": "Hello, how are you?"})
print(res)
```

### User Define Chain Node


```python
def convert_to_uppercase(text: str) -> str:
    return text.upper()
```

```python
chains = prompts | llm | out | convert_to_uppercase    # these all are functions(runnables) which can be used in chain and they return runnable output which can be used in next step of chain
res = chains.invoke({"language": "Spanish", "query": "Hello, how are you?"})
print(res)
```
